```python
for i in range(0,891):
    if(Xdata.Sex[i]=='male'):
        Xdata.Sex[i]=0
    else:
        Xdata.Sex[i]=1
```

```python
for i in range(0,891):
    if(Xdata.Embarked[i]=='S'):
        Xdata.Embarked[i]=1
    elif(Xdata.Embarked[i]=='Q'):
        Xdata.Embarked[i]=2
    else:
        Xdata.Embarked=3
```

```python
for i in range(0,891):
    val=Xdata.Name[i].split(', ')[1].split('.')[0]
    if(val=='Mr'):
        Xdata.Name[i]=1
    elif(val=='Master'):
        Xdata.Name[i]=1.5
    elif(val=='Miss'):
        Xdata.Name[i]=2
    elif(val=='Ms'):
        Xdata.Name[i]=3
    elif(val=='Mrs'):
        Xdata.Name[i]=4
    else:
        Xdata.Name[i]=0
```

```python
model=keras.models.Sequential()
```

```python
model.add(layer=keras.layers.Dense(1,input_shape=(7,),activation='relu'))
model.compile('adam',loss='mse',metrics=['accuracy'])
```

```python
din=keras.layers.Input(shape=(7,))
```

```python
dh1=keras.layers.Activation(activation='relu')(keras.layers.Dense(7)(din))
dh2=keras.layers.Activation(activation='relu')(keras.layers.Dense(7)(dh1))
dh3=keras.layers.Activation(activation='relu')(keras.layers.Dense(7)(dh2))
dh4=keras.layers.Activation(activation='relu')(keras.layers.Dense(7)(dh3))
dh5=keras.layers.Activation(activation='relu')(keras.layers.Dense(7)(dh4))
```

```python
dout=keras.layers.Activation('sigmoid')(keras.layers.Dense(1)(dh5))
```
