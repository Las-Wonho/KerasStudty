```python
def tmp_001(data):
    if(data=='Abnorml'):
        return 1
    elif(data=='Normal'):
        return 2
    elif(data=='Partial'):
        return 3
    else:
        return 0
def isNaN(data):
    if(data==data):
        return data
    else:
        return 2000
```

```python
din=keras.layers.Input(shape=(3,))
```

```python
dh1=keras.layers.Activation('relu')(keras.layers.Dense(9)(din))
dh4=keras.layers.Activation('relu')(keras.layers.Dense(5)(dh1))
dh5=keras.layers.Activation('relu')(keras.layers.Dense(2)(dh4))
```

```python
dout=keras.layers.Activation('softmax')(keras.layers.Dense(1)(dh5))
```

```python
model=keras.models.Model(din,dout)
```

```python
model.compile(loss='mse',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(Xdata,Y,epochs=200)
```
