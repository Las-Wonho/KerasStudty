```python
import keras
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('c:\\Users\\haho6\\Desktop\\flavors_of_cacao.csv')
Y=np.array(data.Rating)
data.head()
```

```python
col=data.columns[4]
cocoa=[]
for i in range(0,1795):
    cocoa.append(int(data[col][i][:2]))
```

```python
been=[]
ty=data.columns[7]
for i in range(0,1795):
    if(data[ty][i]==''):
        been.append(0)
    elif(data[ty][i]=='Criollo'):
        been.append(1)
    elif(data[ty][i]=='Trinitario'):
        been.append(2)
    else:
        been.append(-1)
been
```

```python
X=[]
```

```python
for i in range(0,1795):
    X.append([cocoa[i],data.REF[i],been[i]])   
X
```

```python
model=keras.models.Sequential()
```

```python
model.add(keras.layers.Dense(2,input_shape=(3,)))
```

```python
model.add(keras.layers.Dense(4,activation='relu'))
model.add(keras.layers.Dense(8,activation='relu'))
model.add(keras.layers.Dense(8,activation='relu'))
model.add(keras.layers.Dense(4,activation='relu'))
model.add(keras.layers.Dense(1))
```

```python
model.compile(loss='mse',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(X,Y,epochs=300)
```
